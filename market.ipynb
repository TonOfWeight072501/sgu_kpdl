```python
import pandas as pd
marketing_data = pd.read_csv("marketing_campaign.csv",sep='\t')
marketing_data = marketing_data[['ID','Year_Birth', 'Education',
 'Marital_Status','Income','Kidhome', 'Teenhome',
 'Dt_Customer', 'Recency','NumStorePurchases',
 'NumWebVisitsMonth']]
marketing_data.head(5)
```

```python
has_null = marketing_data.isnull().sum().any()
has_nan  = marketing_data.isna().sum().any()
n_duplicated = marketing_data.duplicated().sum()
print(f'Tính toàn vẹn dữ liệu:')
print(f'+ Có giá trị Null: {has_null}')
if has_null:
    display(marketing_data[marketing_data.isnull().any(axis=1)])
print(f'+ Có giá trị Nan: {has_nan}')
if has_nan:
    display(marketing_data[marketing_data.isna().any(axis=1)])
print(f'+ Số dòng trùng: {n_duplicated}')
if n_duplicated > 0:
    display(marketing_data[marketing_data.duplicated()])
```
