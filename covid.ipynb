```python
import numpy as np
import pandas as pd
from scipy import stats
# Load the .csv into a dataframe using read_csv
covid_data = pd.read_csv("owid-covid-data.csv")
covid_data = covid_data[['iso_code','continent',
 'location','date','total_cases','new_cases']]
# Take a quick look at the data
print("5 dòng đầu tiên của dữ liệu: ")
display(covid_data.head(5))
print("5 dòng cuối cùng của dữ liệu: ")
display(covid_data.tail(5))
covid_data.dtypes
covid_data.shape

```

```python
has_null = covid_data.isnull().sum().any()
has_nan  = covid_data.isna().sum().any()
n_duplicated = covid_data.duplicated().sum()
print(f'Tính toàn vẹn dữ liệu:')
print(f'+ Có giá trị Null: {has_null}')
if has_null:
    display(covid_data[covid_data.isnull().any(axis=1)])
print(f'+ Có giá trị Nan: {has_nan}')
if has_nan:
    display(covid_data[covid_data.isna().any(axis=1)])
print(f'+ Số dòng trùng: {n_duplicated}')
if n_duplicated > 0:
    display(covid_data[covid_data.duplicated()])
```

```python
covid_data_withoutna = covid_data.dropna(how = 'any')
covid_data_withoutna.shape
```

```python

# Get the mean of the data
data_mean = np.mean(covid_data_withoutna["new_cases"])
# Get the median of the data
data_median = np.median(covid_data_withoutna["new_cases"])
# Get the mode of the data
# Lưu ý: stats.mode trả về một đối tượng, cần lấy giá trị mode thực tế
data_mode = stats.mode(covid_data_withoutna["new_cases"])
# Obtain the variance of the data
data_variance = np.var(covid_data_withoutna["new_cases"])
# Obtain the standard deviation of the data
data_sd = np.std(covid_data_withoutna["new_cases"])
# Compute the maximum and minimum values of the data
data_max = np.max(covid_data_withoutna["new_cases"])
data_min = np.min(covid_data_withoutna["new_cases"])
# Obtain the 60th percentile of the data
data_percentile = np.percentile(covid_data_withoutna["new_cases"], 60)
# Obtain the quartiles of the data
data_quartile_q3 = np.quantile(covid_data_withoutna["new_cases"], 0.75)
# Get the IQR of the data
data_IQR = stats.iqr(covid_data_withoutna["new_cases"])

desc = covid_data.describe()

display(desc)

df_statistic = pd.DataFrame({
    "mean": data_mean,
    "median": data_median,
    "mode": data_mode,
    "variance": data_variance,
    "standard deviation": data_sd,
    "max": data_max,
    "min": data_min,
    "data percentile (60th)":data_percentile,
    "data quartile (75%)": data_quartile_q3,
    "IQR": data_IQR
}) 


display(df_statistic.loc[0])

# display(combine)
```
